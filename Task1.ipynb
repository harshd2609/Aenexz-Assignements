```python
#Ques1:Write a program to find the largest of three numbers.
a = int(input('enter the first number:'))
b = int(input('enter the second number:'))
c = int(input('enter the third number:'))
if a>b and a>c:
    print('a is larger')
elif b>a and b>c:
    print('b is larger')
else:
    print('c is greater')
```

```python
#Ques2:Write a program to check whether a given year is a leap year.
year = int(input('enter a year:'))
if (year % 4 == 0 and year % 100 != 0)or(year % 400 == 0):
    print('It is the leap Year')
else:
    print('It is not a leap Year')
```

```python
#Ques3:Write a program to check if a character is a vowel or consonant.
n = input('enter a character:')
if n == 'a' or n == 'e' or n == 'i' or n == '0' or n == 'u':
    print('the character is vowel')
else:
    print('the character is consonant')
```

```python
#Ques4:Write a program to check whether a number is divisible by both 5 and 11.
a = int(input('enter the number:'))
if a%5==0 and a%11==0:
    print('Number is divisible by 5 and 11')
else:
    print('Number is not divisible by 5 and 11')
```

```python
#Ques5:Write a program to calculate the sum of first N natural numbers using a while loop.
n = int(input('enter the number:'))
summ = 0
i = 1
while i<=n:
    print(f'{summ} + {i} = {summ+i}')
    summ = summ+i
    i+=1
```

```python
#Ques6: Write a program to print the multiplication table of a given number using a while loop.
n = int(input('enter the number:'))
i = 1
while i<=10:
    print(f'{n} * {i} = {n*i}')
    i+=1
```

```python
'''Ques7: Print a pyramid pattern:
    *
   ***
  *****
 *******  '''
n = int(input('enter the number: '))
spaces = n - 1

for i in range(1, 2*n, 2):
    print(' ' * spaces + i * '*')
    spaces -= 1
```

```python
'''Ques8: Print:
1
12
123
1234
12345'''
n = int(input('enter the number: '))

for i in range(1, n+1):
    row = ""
    for j in range(1, i+1):
        row += str(j)
    print(row)
```
