```python
#Ques1: Write a function that takes the radius of a circle and returns its area.
def calculate_area(radius):
    pi = 3.14
    area = pi*radius*radius
    return area
```

```python
calculate_area(7)
```

```python
#Ques2: Write a function that accepts three numbers and returns the largest one.
def largest(n,a,b):
    if n>=a and n>=b:
        return n
    elif a>=n and a>=b:
        return a
    else:
        return b
```

```python
print(f'{largest(11,4,7)} is largest')
```

```python
#Ques3: Write a function that takes a string and returns its length without using built-in length functions.
def length(text):
    count = 0 
    for char in text:
        count = count+1
    return count    
```

```python
length('harsh')
```

```python
#Ques4: Write a function that checks whether a number is prime.
def prime(n):
    if n<=1:
        return False
        
    for i in range(2,n):
        if n % i == 0:
            return False
            
    return True       
```

```python
prime(103)
```

```python
#Ques5: Write a recursive function to find the nth Fibonacci number.
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
fib(9)
```

```python
#Ques 6 : Write a recursive function to reverse a string.
def rev_string(text):
    if text == '':
        return text
    else:
        return rev_string(text[1:]) + text[0]
        
```

```python
rev_string('harsh')
```

```python
#Ques7: Write a program to find the sum of all numbers from 1 to n using a for loop.
n = int(input('Enter a number:'))
u = 0
for i in range(1,n+1):
    value = u
    u = u+i
    print(f'{value} + {i} = {u}')
```

```python
#Ques8: Write a program to count how many even numbers are present in a list.
numbers = [1,2,3,4,5,6,7,8,34,56,78,43,21]
count = 0 
for i in numbers:
    if i % 2 ==0:
        print(f'{i} is a even number')
        count = count+1

print('Total even numbers:',count)
    
```
