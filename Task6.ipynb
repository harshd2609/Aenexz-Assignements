```python
from sklearn.datasets import fetch_california_housing
```

```python
cal_housing = fetch_california_housing()
```

```python
cal_housing
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(cal_housing.data , columns = cal_housing.feature_names)
```

```python
df
```

```python
df['Price'] = cal_housing.target 
```

```python
df
```

```python
X = df.drop('Price' , axis = 1)
Y = df['Price']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X ,Y , test_size = 0.2 , random_state = 42)
```

```python
model = LinearRegression()
model.fit(X_train , Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
print("R2 Score:", r2_score(Y_test, y_pred))
```

```python
import matplotlib.pyplot as plt 
plt.figure(figsize=(6,6))
plt.scatter(Y_test, y_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.show()
```

```python
print("Coefficients:", model.coef_)

for feature, coef in zip(X.columns, model.coef_):
    
    if coef > 0:
        print(f"{feature} has a positive relationship with Price.")
        
    elif coef < 0:
        print(f"{feature} has a negative relationship with Price.")
        
    else:
        print(f"{feature} has no relationship with Price.")
```
